#PrediNetworks

##1. Models

###A. Predictive Auto-encoder Rate Model (ARM)

####1. Ideal model (IARM)

Assuming that we want a rate network $\mathbf{r}$ to express an estimate $\mathbf{\hat{x}}$ of a variable 
$\mathbf{x}$, obtained from a linear decoder $\mathbf{D}$ so as $\mathbf{x}=\mathbf{\hat{x}}=\mathbf{D}\mathbf{r}$.
The loss is usually formulated as 
the quadratic difference between true signal and its estimate:

\begin{eqnarray}
L(t)=\frac{1}{2} || \hat{\mathbf{{x}}}(t)-\mathbf{x}(t) ||^2  + \frac{\mu}{2}||\mathbf{r}(t)||^2
\end{eqnarray}

Computing the gradient term leads then to :

\begin{eqnarray}
\dot{\mathbf{r}} &\propto& -\frac{\partial L}{\partial \mathbf{r}} = -\mathbf{D}^{T}( \mathbf{D}\mathbf{r} -\mathbf{x} )-\mu\mathbf{r}\\
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}-\mathbf{D}^{T}\mathbf{D}\mathbf{r} + \mathbf{D}^{T}\mathbf{x}
\end{eqnarray}



####2. Satellite models 

We look at the generic equivalent model :


\begin{eqnarray}
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}-\mathbf{\Omega_{f}}\mathbf{r} + \mathbf{F}\mathbf{x}\\
\mathbf{\hat{x}} &=& \mathbf{D}\mathbf{r}
\end{eqnarray}

#####a. Perturbated model (PeARM) (Thanks to Matty !)

In the perturbative case, lateral and input weights can be modified from the ideal case :

\begin{eqnarray}
\mathbf{F}=\mathbf{D}^T+\mathbf{P_{F}} \\
\mathbf{\Omega_{f}}=\mathbf{D}^T\mathbf{D}+\mathbf{P_{\Omega}}
\end{eqnarray}

where $\mathbf{P_{i}}\propto\mathcal{N}(0,\sigma_{i})$

This gives the framework for the Predictive Chaotic Rate Regime (PCRR)

#####b. Platic model (PlARM)

Lateral weights are learnt with a learning rate $\alpha$ :

\begin{eqnarray}
\dot{\mathbf{\Omega_{f}}}=\alpha(\mathbf{x}\mathbf{r}^{T}-d \mathbf{\Omega_{f}})
\end{eqnarray}

Input weights can also be learnt on a longer time scale $\beta$

\begin{eqnarray}
\dot{\mathbf{F}} = \beta(\mathbf{r}\mathbf{x}^{T}-f \mathbf{F})
\end{eqnarray}

###B. Predictive Dynamic-encoder Rate Model (DRM)

####1. Ideal model (IDRM) (Thanks to Wieland !)


Assuming now that we want a rate network to express an estimate of a variable 
$\mathbf{x}$, obtained from a linear system of differential equations so as : 

\begin{eqnarray}
\dot{\mathbf{x}}(t) = \mathbf{A}\mathbf{x}(t) + \mathbf{c}(t)
\end{eqnarray}

Like in the classic model, let's using a linear decoder $\mathbf{D}$ so as the estimate from 
the network $\mathbf{\hat{x}}$ is obtained from the rate variable $\mathbf{r}$ such as : 

\begin{eqnarray}
\dot{\hat{\mathbf{x}}}(t) = -\lambda \mathbf{\hat{x}}(t)+ \mathbf{D}\mathbf{r}(t)
\end{eqnarray}

where $\lambda$ is the decoder time constant. As Wieland proposes nicely, different to the standard autoencoder, where the loss is usually formulated as the quadratic difference between true signal and its estimate, we here define the quadratic loss on the first order dynamics of the decoded signal. 

\begin{eqnarray}
L(t)&=& \frac{1}{2} || \dot{\hat{\mathbf{{x}}}}(t)-\dot{\mathbf{x}}(t) ||^2  + \frac{\mu}{2}||\mathbf{r}(t)||^2
\end{eqnarray}


Assuming that the system tracks well so as 
$\mathbf{x}(t) \approx \hat{\mathbf{x}}(t) = \mathbf{D}(\mathbf{r}*e^{-\lambda u})(t)=\mathbf{D}\bar{\mathbf{r}}(t)$, 
computing the gradient term leads then to :

\begin{eqnarray}
\dot{\mathbf{r}} &\propto& -\frac{\partial L}{\partial \mathbf{r}} = -\mathbf{D}^{T}( -\lambda \mathbf{\hat{x}}- \mathbf{D}\mathbf{r}-\mathbf{A}\mathbf{x} -\mathbf{c} )-\mu\mathbf{r}\\
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}+\mathbf{D}^{T}(\mathbf{A}+\lambda \mathbf{I})\mathbf{D}\bar{\mathbf{r}} -\mathbf{D}^{T}\mathbf{D}\mathbf{r} + \mathbf{D}^{T}\mathbf{c}
\end{eqnarray}



####2. Satellite models 

As previously, the more generic equivalent model is :

\begin{eqnarray}
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}+\mathbf{\Omega_{s}}\bar{\mathbf{r}} -\mathbf{\Omega_{f}}\mathbf{r} + \mathbf{D}^{T}\mathbf{c}
\end{eqnarray}

This gives the framework for the Predictive Fourier Rate Regime (PFRR)


###C. Predictive Encoder Spike Model (SM)

####1. Ideal model (ISM)


Now the Agents are going to be Integrate and Fire neurons. The decoder then integrates the spike trains so as :

\begin{eqnarray}
\mathbf{\dot{\hat{x}}}=-\lambda_{d}\mathbf{\hat{x}}+\mathbf{D}\mathbf{o}
\end{eqnarray}

The Voltage equation for each LIF is :

\begin{eqnarray}
\dot{\mathbf{V}}=\mathbf{D}^T(\mathbf{A}+\lambda_{d}\mathbf{I})\mathbf{\hat{x}}- \mathbf{D}^T\mathbf{D} \mathbf{o}+ \mathbf{D}^{T} \mathbf{c}
\end{eqnarray}

with $V_{r,i}=\sum {D_{ij}}^2/2$

Note that the SM is like an autoencoder for $\mathbf{A}=-a\mathbf{I}$ and $a=\lambda_{d}$, then this is the ASM model.
(DSM else).



####2. Satellites models

\begin{eqnarray}
\dot{\mathbf{V}}&=& -\lambda_{V}(\mathbf{V}-E_{m})+ \mathbf{\Omega_{s}}\mathbf{\hat{x}} - \mathbf{\Omega_{f}} \mathbf{o} + \mathbf{F} \mathbf{c}\\
\mathbf{\dot{\hat{x}}}&=&-\lambda_{d}\mathbf{\hat{x}}+\mathbf{D}\mathbf{o}
\end{eqnarray}


#####a. Perturbated model (PeASM)

#####b. Platic model (PlASM)

\begin{eqnarray}
\Delta \Omega_{f,ij} &=& \alpha o_{i} ( (V_{i}-E_{m})+ (1/2) \Omega_{f,ij} )\\
\Delta F_{ik} &=& \alpha o_{i} ( \bar{c}_{k} - F_{ik} )
\end{eqnarray}
