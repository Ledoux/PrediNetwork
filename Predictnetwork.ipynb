##1. Chaotic dynamic versus robust predictive decoding

###A. Predictive autoencoder rate model 

Assuming that we want a rate network $\mathbf{r}$ to express an estimate $\mathbf{\hat{x}}$ of a variable 
$\mathbf{x}$, obtained from a linear decoder $\mathbf{D}$ so as $\mathbf{x}=\mathbf{\hat{x}}=\mathbf{D}\mathbf{r}$.
The loss is usually formulated as 
the quadratic difference between true signal and its estimate:

\begin{eqnarray}
L(t)=\frac{1}{2} || \hat{\mathbf{{x}}}(t)-\mathbf{x}(t) ||^2  + \frac{\mu}{2}||\mathbf{r}(t)||^2
\end{eqnarray}

Computing the gradient term leads then to :

\begin{eqnarray}
\dot{\mathbf{r}} &\propto& -\frac{\partial L}{\partial \mathbf{r}} = -\mathbf{D}^{T}( \mathbf{D}\mathbf{r}(t) -\mathbf{x}(t) )-\mu\mathbf{r}(t)\\
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}(t)-\mathbf{D}^{T}\mathbf{D}\mathbf{r}(t) + \mathbf{D}^{T}\mathbf{x}(t)
\end{eqnarray}

Now we are going to add a little perturbation on the lateral connectivity such as :

\begin{eqnarray}
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}(t)-\mathbf{\Omega}\mathbf{r}(t) + \mathbf{D}^{T}\mathbf{x}(t)
\end{eqnarray}

where $\mathbf{\Omega}=\mathbf{D}^T\mathbf{D}+\mathbf{P}$ and $\mathbf{P}\propto\mathcal{N}(0,\sigma)$

#2. Predictive dynamic models 

###A.Rate model 


Assuming now that we want a rate network to express an estimate of a variable 
$\mathbf{x}$, obtained from a linear system of differential equations so as : 

\begin{eqnarray}
\dot{\mathbf{x}}(t) = \mathbf{A}\mathbf{x}(t) + \mathbf{c}(t)
\end{eqnarray}

Like in the classic model, let's using a linear decoder $\mathbf{D}$ so as the estimate from 
the network $\mathbf{\hat{x}}$ is obtained from the rate variable $\mathbf{r}$ such as : 

\begin{eqnarray}
\dot{\hat{\mathbf{x}}}(t) = -\lambda \mathbf{\hat{x}}(t)+ \mathbf{D}\mathbf{r}(t)
\end{eqnarray}

where $\lambda$ is the decoder time constant. As Wieland proposes nicely, different to the standard autoencoder, where the loss is usually formulated as the quadratic difference between true signal and its estimate, we here define the quadratic loss on the first order dynamics of the decoded signal. 

\begin{eqnarray}
L(t)&=& \frac{1}{2} || \dot{\hat{\mathbf{{x}}}}(t)-\dot{\mathbf{x}}(t) ||^2  + \frac{\mu}{2}||\mathbf{r}(t)||^2
\end{eqnarray}


Assuming that the system tracks well so as 
$\mathbf{x}(t) \approx \hat{\mathbf{x}}(t) = \mathbf{D}(\mathbf{r}*e^{-\lambda u})(t)=\mathbf{D}\bar{\mathbf{r}}(t)$, 
computing the gradient term leads then to :

\begin{eqnarray}
\dot{\mathbf{r}} &\propto& -\frac{\partial L}{\partial \mathbf{r}} = -\mathbf{D}^{T}( -\lambda \mathbf{\hat{x}}(t)- \mathbf{D}\mathbf{r}(t)-\mathbf{A}\mathbf{x}(t) -\mathbf{c}(t) )-\mu\mathbf{r}(t)\\
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}(t)+\mathbf{D}^{T}(\mathbf{A}+\lambda \mathbf{I})\mathbf{D}\bar{\mathbf{r}} -\mathbf{D}^{T}\mathbf{D}\mathbf{r}(t) + \mathbf{D}^{T}\mathbf{c}(t)
\end{eqnarray}

As previously, we can add a perturbation matrix on the slow and the fast current, the model will be then

\begin{eqnarray}
\dot{\mathbf{r}}  &=& -\mu\mathbf{r}(t)+\Omega_{s}\bar{\mathbf{r}} -\Omega_{f}\mathbf{r}(t) + \mathbf{D}^{T}\mathbf{c}(t)
\end{eqnarray}

where $\mathbf{\Omega}_{s}=\mathbf{D}^T(\mathbf{A}+\lambda \mathbf{I})\mathbf{D}+\mathbf{P}_{s}$
and $\mathbf{\Omega}_{f}=\mathbf{D}^T\mathbf{D}+\mathbf{P}_{f}$ 

with $\mathbf{P}_{i}\propto\mathcal{N}(0,\sigma_{i})$

###B. Spike model 


Now the Agents are going to be Integrate and Fire neurons. The decoder then integrates the spike trains so as :

\begin{eqnarray}
\mathbf{\dot{\hat{x}}}(t)=-\lambda_{d}\mathbf{\hat{x}}(t)+\mathbf{D}\mathbf{o}(t)
\end{eqnarray}

The Voltage equation for each LIF is :

\begin{eqnarray}
\dot{\mathbf{V}}=\mathbf{D}^T(\mathbf{A}+\lambda_{d}\mathbf{I})\mathbf{\hat{x}}(t) - \mathbf{D}^T\mathbf{D} \mathbf{o(t)} + \mathbf{D}^{T} \mathbf{c(t)}
\end{eqnarray}

with $\mathbf{V_{r}}=\mathbf{D}^2/2$
